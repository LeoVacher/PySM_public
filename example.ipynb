```python
import pysm
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

# Initialization
Initialize either component object, or sky object, from model strings.

```python
nside = 128
mbb = pysm.preset_models('d1', nside)
sky = pysm.Sky(nside, preset_strings=['d1'])
```

```python
frequencies = np.array([15., 150., 400.])
bandpasses = [(np.linspace(f-1, f+1, 50), np.ones(50)) for f in [15., 150., 400.]]
fwhms = np.array([120., 30., 10.])
```

# Functionality
Functionality of Model objects consists of three main methods: 

- `Model.get_emission(frequencies)`
- `Model.apply_bandpass(bandpasses)`
- `Model.apply_smoothing(data, fwhms)`

Since `Sky` is subclassed from `Model`, these are all present when defining a sky model composed of a group of components through e.g. `pysm.Sky(nside, preset_strings=['d1', 's1', 'a1'])`.

```python
mbb_out = mbb.get_emission(frequencies)
mbb_out_bpass = mbb.apply_bandpass(bandpasses)
mbb_out_smoothed = mbb.apply_smoothing(mbb_out, fwhms)
mbb_out_bpass_smoothed = mbb.apply_smoothing(mbb_out, fwhms)
```

```python
hp.mollview(mbb_out[0, 0], norm='log', min=0.1, max=50)
hp.mollview(mbb_out_bpass[0, 0], norm='log', min=0.1, max=50)
```
